```python
salary_data=pd.read_csv(input_file_path)
salary_data.describe()
```

```python
x=salary_data["YearsExperience"]
y=salary_data["Salary"]
salary_data
```

```python
plt.scatter(x,y)
plt.show()
```

```python
w=0 #Initial Weight
b=0 #Intial Bias
l=0.0053 #Initial Learning Rate
turns=10000 #No. of Epochs
```

```python
#Gradient Descent Function
def gradient(w1,b1,points,L):
    w_grad=0
    b_grad=0
    n = len(points)
    for i in range(n):
        x=points.iloc[i].YearsExperience #what iloc does is basically its taking the value from Years_Experience and salary based on the index
        y=points.iloc[i].Salary
        w_grad+=-(2/n)*x*(y-(w1*x+b1))
        b_grad+=-(2/n)*(y-(w1*x+b1))

    w_new=w1-w_grad*L
    b_new=b1-b_grad*L
    return w_new,b_new
```

```python
#Plotting the CSV file initial Data
fig,ax=plt.subplots()
ax.scatter(x,y,color="blue")
ax.set_xlabel("Years_Experience")
ax.set_ylabel("Salary")
ax.set_title("Gradient Descent")
line=ax.plot([], [],color="red",linewidth=2)
```

```python
#Plotting for each epoch that is divisible by 1000
for epoch in range(turns):
    w,b=gradient(w,b,salary_data,l)
    if epoch%1000==0:
        plt.figure(figsize=(8,6))
        plt.scatter(salary_data.YearsExperience,salary_data.Salary,label="Data Points")
        x_vals=salary_data.YearsExperience
        y_vals=w*(x_vals)+b
        plt.plot(x_vals,y_vals,color="red",label=f"Epoch {epoch}: y={w:.2f}x + {b:.2f}")
        plt.legend()
        plt.title(f"Linear fit at Epoch {epoch}")
        plt.xlabel("Years of Experience")
        plt.ylabel("Salary")
        plt.show()
        
        
        
        
    
```
